# Data Exploration Assignment Part 2 - Ann Arbor Water Mains Data

```python
# Load libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Directly import the watermain breaks csv from the Google Drive data inventory
url_watermain_breaks = 'https://drive.google.com/file/d/1LTYIn0aq4gBmD65B508miLt8cSzfLom8/view?usp=sharing'
url_download_watermain_breaks ='https://drive.google.com/uc?export=download&id=' + url_watermain_breaks.split('/')[-2]
watermain_breaks_data = pd.read_csv(url_download_watermain_breaks)
watermain_breaks_data
```

```python
# Directly import the watermain attributes csv from the Google Drive data inventory
url_watermain_attributes = 'https://drive.google.com/file/d/1i0jouGqBSRERM8yKp1jeQTKo5RWKZO5I/view?usp=sharing'
url_download_watermain_attributes ='https://drive.google.com/uc?export=download&id=' + url_watermain_attributes.split('/')[-2]
watermain_attributes_data = pd.read_csv('https://docs.google.com/spreadsheets/d/1pIrZNiLOfzLmbNErw2JdtS0X0exFCEPLYovPrTzEauk/gviz/tq?tqx=out:csv')
watermain_attributes_data
```

Question 1:

```python
#get data only in the range 1/1/2018 to 7/31/2023
watermain_breaks_data['Actual Start'] = pd.to_datetime(watermain_breaks_data['Actual Start'])
start_date = '1/1/2018'
end_date = '7/31/2023'
watermain_breaks_data = watermain_breaks_data[(watermain_breaks_data['Actual Start'] >= start_date) & (watermain_breaks_data['Actual Start'] <= end_date)]
```

```python
#get average number of watermain breaks per month
watermain_breaks_data['Months'] = watermain_breaks_data['Actual Start'].dt.month
breaks_per_month = watermain_breaks_data.groupby('Months')['Work Order ID'].count()
highest_breaks_per_month = breaks_per_month.max()
print('The month with the highest number of watermain breaks is', highest_breaks_per_month, '.')
```

```python
#show data
breaks_per_month
```

Month 11, or November, has the highest number of watermain breaks on average.

```python
#plot monthly averages
plt.title('Average watermain breaks per month')
plt.xlabel('Months')
plt.ylabel('Number of watermain breaks')
plt.plot(breaks_per_month)
plt.show()
```

There seem to be less watermain breaks during the middle of summer months of the year, with higher averages during winter months January, February, November, and December. There is a slightly higher average number of watermain breaks in July than other summer months, but no other large differences.

Question 2:

```python
#want to find number of cancelled orders

#prepare data
#get years from the data start dates
watermain_breaks_data['Years'] = watermain_breaks_data['Actual Start'].dt.year
#replace canceled Y/N with 1 or 0
watermain_breaks_data['Canceled'] = watermain_breaks_data['Canceled'].str.replace('Y', '1')
watermain_breaks_data['Canceled'] = watermain_breaks_data['Canceled'].str.replace('N', '0')
```

```python
#get the average number of canceled orders per month and year
work_orders = watermain_breaks_data.groupby(['Years','Months'])['Canceled'].mean()
#convert into a list to easily plot
canceled = list(work_orders)
```

```python
#plot cancellation rate
plt.title('Order Cancellation Rate vs. Months')
plt.xlabel('Months')
plt.ylabel('Work Order Cancellation Rate')
plt.plot(canceled)
plt.show()
```

There were many cancellations during the inital few months, a few during about month 25, and none otherwise.

Question 3:

```python
#merge data on work order ID
#breaks ID: 'Work Order ID'
#attributes ID: 'WORKORDERI'

#rename watermain attributes column so that the column names match
#and the data sets can be merged on that column name
watermain_attributes_data.rename(columns={'WORKORDERI': 'Work Order ID'}, inplace=True)
watermain_merged = pd.merge(watermain_breaks_data, watermain_attributes_data, on='Work Order ID')
watermain_merged
```

```python
#get average amount of watermains per work order
work_orders = watermain_merged.groupby('Work Order ID')['Work Order ID'].count()
avg_per_order = work_orders.mean()
print('The average number of watermains associated with a single work order is', avg_per_order, '.')
```

Question 4:

```python
#for each work order
#compute lengths
order_lengths = watermain_merged.groupby('Work Order ID')['LENGTH'].sum()
#show lengths data to see what they look like
order_lengths
```

```python
#convert start and finish times to datetime objects
watermain_merged['Actual Start'] = pd.to_datetime(watermain_merged['Actual Start'])
watermain_merged['Actual Finish'] = pd.to_datetime(watermain_merged['Actual Finish'])
#get the amount of time spent on each work order
watermain_merged['Total Time'] = watermain_merged['Actual Finish'].dt.day - watermain_merged['Actual Start'].dt.day
```

```python
#get the total time spent on each work order
times_by_order = watermain_merged.groupby('Work Order ID')['Total Time'].sum()
#if there are any NaN values, the order took less than a day
#fill NaNs with 0, as it took 0 days
times_by_order = times_by_order.fillna(0)
#get int values to graph values
times_by_order.transform(lambda x: x.astype(int))
#look at the distribution of the data
times_by_order.describe()
```

```python
#get rid of the negative values, likely bad data
data_valid = watermain_merged[(watermain_merged['Total Time'] >= 0)]
times_by_order_valid = data_valid.groupby('Work Order ID')['Total Time'].sum()
#look at the data again
times_by_order_valid.describe()
```

```python
#make sure to remove them in order_lengths as well
order_lengths_valid = data_valid.groupby('Work Order ID')['LENGTH'].sum()
#look at the data again
order_lengths_valid.describe()
```

```python
#no more negative values
#plot lengths and finish times
plt.title('Time to Finish Work Orders vs. Lengths of Watermains')
plt.xlabel('Lengths (ft.)')
plt.ylabel('Time to finish Work Order (days)')
plt.scatter(order_lengths_valid, times_by_order_valid)
plt.show()
```

Most watermains took less than a day to complete, but shorter watermains had some work orders that took multiple days to complete. These were mostly between 1-3 days, with one possible outlier taking 7 days. Watermains between roughly 0 and 600 feet had some variation in work order time length, again mostly varying between 1 to 3 days. There are no watermains over 600 feet that took more than a day to finish. This could be because longer watermains are given more support so that the work orders are done more quickly, and smaller watermains are less important and can take more time to work on.
